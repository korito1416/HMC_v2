```python
import numpy as np
from scipy import integrate
from scipy.stats import gaussian_kde

# Generate the KDE functions
kde1 = gaussian_kde(theta_samples)
kde2 = gaussian_kde(results['collected_ensembles'][i][:, j])

# Numerically integrate from 0 to infinity
int1, _ = integrate.quad(kde1.evaluate, 0, np.inf)
int2, _ = integrate.quad(kde2.evaluate, 0, np.inf)

# Print the results
print(f"Integration result for sample1: {int1}")
print(f"Integration result for sample2: {int2}")
```

```python
# Define the range for the x-axis
x_range = np.linspace(0, np.max([theta_samples, results['collected_ensembles'][i][:, j]]), 500)

# Compute the densities
density1 = kde1.evaluate(x_range)
density2 = kde2.evaluate(x_range)

# Create a new figure
plt.figure(figsize=(10,6))

# Plot the KDE for sample1
plt.plot(x_range, density1, label='Sample 1', alpha=0.7)

# Plot the KDE for sample2
plt.plot(x_range, density2, label='Sample 2', alpha=0.7)

# Set the title
plt.title('Density Plots using scipy')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Assume these are your samples (truncated at 0)
sample1 = theta_samples
sample2 = results['collected_ensembles'][i][:, j]

# Generate the KDE functions
kde1 = gaussian_kde(sample1)
kde2 = gaussian_kde(sample2)

# Define the range for the x-axis
x_range = np.linspace(0, np.max([sample1, sample2]), 500)

# Compute the densities
density1 = kde1.evaluate(x_range)
density2 = kde2.evaluate(x_range)

# Create a figure with 2 subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(16,6))

# --- Subplot 1: Density plots using scipy ---
axes[0].plot(x_range, density1, label='Sample 1 - scipy', alpha=0.7)
axes[0].plot(x_range, density2, label='Sample 2 - scipy', alpha=0.7)
axes[0].set_title('Density Plots using scipy')
axes[0].legend()

# --- Subplot 2: Density plots using seaborn ---
sns.kdeplot(sample1, ax=axes[1], label='Sample 1 - seaborn', shade=True, clip=(0, None))
sns.kdeplot(sample2, ax=axes[1], label='Sample 2 - seaborn', shade=True, clip=(0, None))
axes[1].set_title('Density Plots using seaborn')
axes[1].legend()

# Display the plot
plt.show()
```
